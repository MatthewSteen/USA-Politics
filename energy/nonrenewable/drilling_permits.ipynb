# Drilling Permits

Oil and gas drilling on Federal lands has become a divisive political issue in the context of gasoline prices, energy security, and climate change. When gasoline prices increase, there's often renewed attention to increasing production from Federal lands. However, this solution ignores several facts. First, crude oil prices are determined by the global market[^1]. Second, the majority of oil production in the US comes from private, not public, land[^2] [^3]. Third, although the US as a whole is a top petroleum producer, the percentage of global production from Federal lands is small because it's a small percentage of total US production.

The purpose of this analysis is to compare the total number of approved drilling permits on Federal land during the 45th and 46th President Of The United States (POTUS) first term.

## Methodology

This analysis uses raw data from the Bureau of Land Management Approved Applications for Permits to Drill (APDs) Report[^4] using the following steps.

1. Set `Date Limit Between` (e.g. 01/20/2027 - 01/19/2020)
2. Click `Spreadsheet Listing` tab
3. Click `Apply` button and wait to download the .xls file
4. Open .xls file and Save As .csv

[^1]: https://www.eia.gov/finance/markets/crudeoil/index.php
[^2]: https://www.eia.gov/energyexplained/oil-and-petroleum-products/where-our-oil-comes-from.php
[^3]: https://crsreports.congress.gov/product/pdf/R/R42432
[^4]: https://reports.blm.gov/report/AFMSS/81/Approved-APDs-Report-Federal

```python
# import python libraries
import pandas as pd
```

```python
# directory of raw data
dir = 'https://raw.githubusercontent.com/MatthewSteen/USA-Politics/energy/drilling-permits/energy/nonrenewable/data/'

# prefix and suffix of files
prefix = 'ApprovedFederalAPDs_'
suffix45 = '20170120-20200119.csv'
suffix46 = '20210120-20240119.csv'

# paths to files
path45 = dir + prefix + suffix45
path46 = dir + prefix + suffix46

# plot arguments
kind = 'bar'
title = 'Drilling Permits'
xlabel = 'Term 1'
ylabel = 'Approved Applications for Permits to Drill'
rot = 0 # rotation

# x axis
term_year = ['Year 1', 'Year 2', 'Year 3', 'Year 4']
```

### POTUS 45

```python
# read data to dataframe
potus45 = pd.read_csv(path45)
potus45
```

```python
# change approval date to datetime format
potus45['APD Approval Date'] = pd.to_datetime(potus45['APD Approval Date'])

# set index to approval date
potus45.set_index(potus45['APD Approval Date'], drop=True, inplace=True)

# sort index
potus45.sort_index(inplace=True)

# show
potus45
```

```python
# https://stackoverflow.com/questions/26646191/pandas-groupby-month-and-year

# group by year
potus45_permits = potus45.groupby([(potus45.index.year)], as_index=True).count()

# drop columns
potus45_permits.drop(potus45_permits.columns[1:], axis=1, inplace=True)

# rename column
potus45_permits.rename(columns={'Admin State': 'Trump'}, inplace=True)

# show
potus45_permits
```

```python
# add year column
potus45_permits[xlabel] = pd.Series(term_year, index=potus45_permits.index)

# set index to year
potus45_permits.set_index(potus45_permits[xlabel], drop=True, inplace=True)

# drop year column
potus45_permits.drop([xlabel], axis=1, inplace=True)

# show
potus45_permits
```

```python
# plot
potus45_permits.plot(kind=kind, title=title, xlabel=xlabel, ylabel=ylabel, rot=rot, color='red')
```

### POTUS 46

```python
# read data to dataframe
potus46 = pd.read_csv(path46)
potus46
```

```python
# change approval date to datetime format
potus46['APD Approval Date'] = pd.to_datetime(potus46['APD Approval Date'])

# set index to approval date
potus46.set_index(potus46['APD Approval Date'], drop=True, inplace=True)

# sort index
potus46.sort_index(inplace=True)

# show
potus46
```

```python
# https://stackoverflow.com/questions/26646191/pandas-groupby-month-and-year

# group by year
potus46_permits = potus46.groupby([(potus46.index.year)], as_index=True).count()

# drop columns
potus46_permits.drop(potus46_permits.columns[1:], axis=1, inplace=True)

# rename column
potus46_permits.rename(columns={'Admin State': 'Biden'}, inplace=True)

# show
potus46_permits
```

```python
# add year column
potus46_permits[xlabel] = pd.Series(term_year, index=potus46_permits.index)

# set index to year
potus46_permits.set_index(potus46_permits[xlabel], drop=True, inplace=True)

# drop year column
potus46_permits.drop([xlabel], axis=1, inplace=True)

# show
potus46_permits
```

```python
# plot
potus46_permits.plot(kind=kind, title=title, xlabel=xlabel, ylabel=ylabel, rot=rot, color='blue')
```

## Results

```python
# data dictionary
data_dict = {
    'Trump': potus45_permits['Trump'],
    'Biden': potus46_permits['Biden']
}

# new dataframe of combined data
permits = pd.DataFrame(data=data_dict)

# show
permits
```

```python
# calculate total
permits.sum(axis=0)
```

```python
# plot
axes_subplot = permits.plot(kind=kind, title=title, xlabel=xlabel, ylabel=ylabel, rot=rot, color=['red', 'blue'])

# figure
figure = axes_subplot.get_figure()
```

## Conclusions

The results above show that there have been 6% more drilling permits approved during Biden's first three years (9963 total) than Trump's first three years (9363 total) in office. These numbers align closely with another analysis of drilling permits approved on federal lands[^2] using a similar data source.
